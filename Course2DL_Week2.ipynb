# Mini-batch gradient descent
- want fast optimization algorithms
- Vectorization allows to efficiently comput on m examples
- $$ X = [x^{(1)} x^{(2)} x^{(3)} ... x^{(m)}] dim = (n_{(x)},m) $$
- $$ Y = [y^{(1)} y^{(2)} y^{(3)} ... y^{(m)}] dim = (1,m) $$
- What if n = 5,000,000 could use minibatches of 1,000 each with $$ X^{'{5000}'}$$
- use round brackets for training examples, square breackets for layers, and curly brackets (using '' because curly does not work in markdown) for mini batch

To run
- for t = 1,..., 5000
1 step of gradient descent
- Forward prop on $$ X^{'{t}'}$$
$$ Z^{(1)} = w^{(1)}X + b^{(1)}$$
$$ A^{[1]} = g^{[1]}(Z^{[1]}) $$
...
$$ A^{[L]} = g^{[L]}(Z^{[L]}) $$
Processes 1000 at a time
Compute cost 
- $$ J(w,b) = \frac{1}{1000}\sum_{i=1}^{\infty}L(\hat{y}^{(i)}, y^{(i)}) + \frac{\lambda}{2*1000}\lVert \mathbf{w} \rVert^{2}_2$$
- Backprop to compute gradient wrt $$ J^{'{t}'}$$ with $$ X^{'{t}'}$$ and $$ Y^{'{t}'}$$
- does one epoch of training, single pass thorugh the training set, so all 5,000 examples
- you might want more passes through training set, so will repeat all mini batch

- compute exponentially weighted average of your gradients, use that gradient to update your weights instead

- to prevent oscilattions, use smalller learing rate

- Compute dW, dV on current mini batch

- $$ V_{dw} =\beta V_{dw} + (1-\beta)dw $$

- $$ w = w- \alpha V_{dw} $$
- $$ b = b- \alpha V_{db} $$
- makes much smaller oscillations, takes a more straight forward path
- now have hyperparameters alpha and beta, beta = 0.9 works very well

# RMSprop

- $$ w = w- \alpha \frac{dw}{\sqrt{S_{dw}}} $$
- $$ b = b- \alpha \frac{db}{\sqrt{S_{db}}} $$

# Adam optimization algortihm
- initalize with
- $$v_{dw} = 0 , S_{dw} = 0, v_{db} = 0 , S_{db} = 0 $$
on iteration t:
- compute dw, db using mini-batch
- momentum like with beta 1
- $$ V_{dw} =\beta_1 V_{dw} + (1-\beta_1)dw $$
- $$ V_{db} =\beta_1 V_{db} + (1-\beta_1)db $$
- RMS prop
- $$ S_{dw} =\beta_2 S_{dw} + (1-\beta_2)dw^2 $$
- $$ S_{db} =\beta_2 S_{db} + (1-\beta_2)db^2 $$
- to correct for bias
- $$ V_{dw}^{corrected} = \frac{V_{dw}}{1- \beta_1^{t}}  $$
- $$ V_{db}^{corrected} = \frac{V_{db}}{1- \beta_1^{t}}  $$
- $$ S_{dw}^{corrected} = \frac{S_{dw}}{1- \beta_2^{t}}  $$
- $$ S_{db}^{corrected} = \frac{S_{db}}{1- \beta_2^{t}}  $$
- $$ w = w- \alpha \frac{V_{dw}^{corrected}}{\sqrt{S_{dw}^{corrected}} + \epsilon} $$
- $$ b = b- \alpha \frac{V_{db}^{corrected}}{\sqrt{S_{db}^{corrected}} + \epsilon} $$
- many hyperparameters : alpha,needs to be tunes, beta1 ; 0.9, beta2: 0.999, epsilon: 10^-8
- use default values of betas and epsilon and tune alpha
- Adam : Adaptive Moment Estimation
- beta1 gets first moment and beta2 gets the second moment

# Learning rate decay
- slowly reduce your learning rate over time
- if using mini batch gradient descent, wont converge bc using a fixed value for alpha
- if slowly reduce alpha, steps are smaller, will oscillate in a tigher region near the minimum
- recall 1 epoch = 1 pass through the data
- - set alpha to
- $$ \alpha = \frac {1}{1 +decayrate *epoch num} \alpha_0 $$
- as epoch number increases, the learning rate gradually decreases
- try a variety of hyperparameters alpha0 and decaying rate

-Other learning rate decay methods
- exponential decay
- $$ \alpha = 0.95^{epoch num}* \alpha_0 $$
- other methods
- $$ \alpha = \frac{k}{\sqrt{epoch num}} * \alpha_0 $$
