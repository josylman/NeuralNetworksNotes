# Forward Propagation
- Layer 1 : $$z^{[1]}= w^{[1]}a^{[0]} + b^{[1]} $$
 $$a^{[1]}= g^{[1]}(z^{[1]}) $$
 Remember : X = $$A^{[0]}$$
 
- Layer 2 : $$z^{[2]}= w^{[2]}a^{[1]} + b^{[2]} $$
 $$a^{[2]}= g^{[2]}(z^{[2]}) $$
 
 - Layer 4 : $$z^{[4]}= w^{[4]}a^{[3]} + b^{[4]} $$
 $$a^{[4]}= g^{[4]}(z^{[4]}) $$
 
 - General for any layer: $$z^{[l]}= w^{[l]}a^{[l-1]} + b^{[l]} $$
 $$a^{[l]}= g^{[l]}(z^{[l]}) $$
 
 - For Vectorized, stacking X, Z, A in columns for training examples
- Layer 1 : $$Z^{[1]}= w^{[1]}A^{[0]} + b^{[1]} $$
 $$A^{[1]}= g^{[1]}(Z^{[1]}) $$
 
- Layer 2 : $$Z^{[2]}= w^{[2]}A^{[1]} + b^{[2]} $$
 $$A^{[2]}= g^{[2]}(Z^{[2]}) $$
 
 - Vectorized propagation, it looks like there will be for loop for each layer, but we have to have a for loop that computes activations for each of the layers

# Backward Propagation

 - General for any layer: $$dz^{[l]}= da^{[l]}*g^{[l]'}(z^{[l]}) $$
 -$$dw^{[l]}= dz^{[l]}a^{[l-1]} $$
  -$$db^{[l]}= dz^{[l]} $$
  -$$da^{[l-1]}= w^{[l]T}dz^{[l]} $$
 
 - Vectorized Version
 -$$dZ^{[l]}= dA^{[l]}*g^{[l]'}(Z^{[l]}) $$
 -$$dw^{[l]}= dZ^{[l]}A^{[l-1]} $$
  -$$db^{[l]}= \frac{1}{m}np.sum(dZ^{[l]}, axis =1, keepdims = True) $$
  -$$dA^{[l-1]}= w^{[l]T}dZ^{[l]} $$
  
  **Summary**
  -First layer, second layer may have Relu and then last has sigmoid, compute the loss, then do back propagation and compute the derivatives
  - would input data x for forward recursion
  - Use the following formula for da
  -$$da^{[l]}= \frac{-y}{a} + \frac{1-y}{1-a}$$
